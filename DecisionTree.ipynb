```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
url = 'https://raw.githubusercontent.com/rashakil-ds/Feature-Engineering/refs/heads/main/customer_purchase_dataset.csv'
```

```python
df = pd.read_csv(url)
```

Preprocessing

```python
df
```

Drop Column

```python
df = df.drop(columns=['Customer_ID'], axis=1)
```

```python
df
```

🔷 STEP 2: Handle Missing Values (Age)
Option 1: Median fill (Best practice)

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

🔷 STEP 3: Encode Target Variable

```python
df['Purchased'] = df['Purchased'].map({'Yes':1,'No':0})
```

```python
df['Purchased']
```

**🔷 STEP 4: Encode Categorical Features**

Categorical columns:

```python
categori_column = ['Gender','Car_Type','Marital_Status',
    'Has_Children','Education']
```

Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
for col in categori_column:
  df[col] = le.fit_transform(df[col])
```

```python
df
```

🔷 STEP 5: Feature / Target split

```python
X = df.drop('Purchased', axis = 1)
X
```

```python
Y = df['Purchased']
Y
```

🔷 STEP 6: Train - Test split

```python
from sklearn.model_selection import train_test_split as tts
```

```python
X_train, X_text, Y_train, Y_test = tts(X,Y, test_size=0.2, random_state=42, stratify = Y) #for yes/no ratio maintain
```

```python
from sklearn.tree import DecisionTreeClassifier
```

One-line Summary

criterion → কীভাবে split করবে

max_depth → গাছ কত বড়

min_samples_split → কত data হলে split

random_state → reproducibility

```python
dt = DecisionTreeClassifier(
    criterion='entropy',   # Information Gain
    max_depth=10,           # Overfitting control
    min_samples_split=40,  # Noise control
    random_state=42
)
```

```python
dt.fit(X_train, Y_train)
```

🔷 5. MODEL EVALUATION

```python
from sklearn.metrics import accuracy_score
y_pred = dt.predict(X_text)
accuracy_score(Y_test, y_pred)
```
