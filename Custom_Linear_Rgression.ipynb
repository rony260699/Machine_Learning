```python
n = len(df)
n
```

```python
dldm = (-2/n)*np.sum(errors*df['Starting (USD)'])
dldc = (-2/n)*np.sum(errors)
print(dldm)
print(dldc)
```

```python
#now update m & c
a = 0.01 #α
m = m - a*dldm
print(m)

c = c - a*dldc
print(c)
```

uporer ta just ekbarer flow. kintu eta kora lagbe totokkhn obdi jotokkhon loss sosjho shimar moddhe na ashe. tolerance = 1e-6         # Stop when loss change is less than this. amra ekhon etake function er akare korbo.

```python
def loss_calculation(df, m, c):
  predics = m*df['Starting (USD)'] + c
  errors = predics - df['Ending (USD)']
  loss = np.mean(errors**2)
  return loss
```

```python
def update_m_c(df, m, c, a):
  error = (m*df['Starting (USD)'] + c) - df['Ending (USD)']
  dldm = (-2/n)*np.sum(df['Starting (USD)']*error)
  dldc = (-2/n)*np.sum(error)

  updated_m = m - a*dldm
  updated_c = c - a*dldc
  return updated_m, updated_c
```

```python
def pred(x, df):
  w = 0
  b = 0
  previous_loss = 0
  a=0.000000001 # Reduced learning rate significantly
  updated_m, updated_c = update_m_c(df, w, b, a)
  updated_loss = 0
  tolerance = 1e-6
  i = 0
  while i < 100000: # Increased iterations to allow for slower convergence
    i += 1
    previous_loss = loss_calculation(df, w, b)
    updated_loss = loss_calculation(df, updated_m, updated_c)
    #print(f"Iteration {i}, Loss: {updated_loss}") # Uncomment to see loss progress
    if(abs(previous_loss - updated_loss) < tolerance):
      print('done')
      break
    w = updated_m
    b = updated_c
    updated_m, updated_c = update_m_c(df, w, b, a)

  predict = updated_m*x+updated_c
  return predict
```

```python
pr = pred(20000, df)
print(pr)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Create a copy of the dataframe to store scaled values
df_scaled = df.copy()

# Initialize scalers for features (X) and target (y)
scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()

# Scale 'Starting (USD)' column
df_scaled['Starting (USD)'] = scaler_X.fit_transform(df_scaled[['Starting (USD)']])

# Scale 'Ending (USD)' column
df_scaled['Ending (USD)'] = scaler_y.fit_transform(df_scaled[['Ending (USD)']])

print("Scaled DataFrame head:")
display(df_scaled.head())
```

Now, we will use the `pred` function with the scaled data and a scaled version of our input `20000`.

```python
# The input value (20000) also needs to be scaled using the same scaler fitted on 'Starting (USD)'
scaled_input_x = scaler_X.transform([[15800]])[0][0]

# Call the prediction function with the scaled input and scaled DataFrame
pr_scaled = pred(scaled_input_x, df_scaled)

print(f"Predicted value on scaled data: {pr_scaled}")
```

Finally, we inverse transform the scaled prediction to get the result in the original USD scale.

```python
# Inverse transform the scaled prediction to get the result in the original scale
pr_original_scale = scaler_y.inverse_transform([[pr_scaled]])[0][0]

print(pr_original_scale)
```
